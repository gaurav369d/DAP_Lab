**Reasoning**:
Implement a function to calculate the sum of digits of an integer using a while loop with arithmetic operations.



````python
def sum_of_digits(n):
    """Calculates the sum of digits using arithmetic operations."""
    n = abs(n)
    total_sum = 0
    while n > 0:
        total_sum += n % 10
        n //= 10
    return total_sum

# Test the function
test_num = 12345
print(f"Sum of digits for {test_num}: {sum_of_digits(test_num)}")
````

**Reasoning**:
Implement a function to calculate the product of digits of an integer using arithmetic operations and a while loop.



````python
def product_of_digits(n):
    """Calculates the product of digits using arithmetic operations."""
    n = abs(n)
    if n == 0:
        return 0
    product = 1
    while n > 0:
        product *= (n % 10)
        n //= 10
    return product

# Test the function
test_num_prod = 1234
print(f"Product of digits for {test_num_prod}: {product_of_digits(test_num_prod)}")
````

Implement a function to reverse an integer using arithmetic operations (extracting digits and building the reversed number) without string conversion as per the third instruction of the subtask.



````python
def reverse_number(n):
    """Reverses the digits of an integer using arithmetic operations."""
    is_negative = n < 0
    n = abs(n)
    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
    return -reversed_num if is_negative else reversed_num

# Test the function
test_num_rev = 12345
print(f"Reversed number of {test_num_rev}: {reverse_number(test_num_rev)}")
````

````python
def is_palindrome(n):
    """Checks if a number is a palindrome using arithmetic operations."""
    if n < 0:
        return False

    original_n = n
    reversed_num = 0
    temp_n = n
    while temp_n > 0:
        digit = temp_n % 10
        reversed_num = reversed_num * 10 + digit
        temp_n //= 10

    return original_n == reversed_num

# Test the function
test_num_pal = 12321
print(f"Is {test_num_pal} a palindrome? {is_palindrome(test_num_pal)}")
test_num_not_pal = 12345
print(f"Is {test_num_not_pal} a palindrome? {is_palindrome(test_num_not_pal)}")
````

Implement the Greatest Common Divisor (GCD) using the iterative Euclidean algorithm in a self-contained code cell.



````python
def gcd(a, b):
    """Calculates the GCD of two numbers using the iterative Euclidean algorithm."""
    a, b = abs(a), abs(b)
    while b:
        a, b = b, a % b
    return a

# Test the function
num1, num2 = 48, 18
print(f"GCD of {num1} and {num2} is: {gcd(num1, num2)}")
````

````python
def lcm(a, b):
    """Calculates the LCM of two numbers using the relationship LCM(a, b) = |a*b| / GCD(a, b)."""
    def gcd_internal(x, y):
        x, y = abs(x), abs(y)
        while y:
            x, y = y, x % y
        return x

    if a == 0 or b == 0:
        return 0

    return abs(a * b) // gcd_internal(a, b)

# Test the function
num1_lcm, num2_lcm = 12, 18
print(f"LCM of {num1_lcm} and {num2_lcm} is: {lcm(num1_lcm, num2_lcm)}")
````

Implement modular exponentiation using the binary exponentiation (square-and-multiply) algorithm. This method is efficient for large exponents as it reduces the number of multiplications and keeps the intermediate results small by applying the modulo at each step.

**Reasoning**:
Implement modular exponentiation using the binary exponentiation (square-and-multiply) algorithm in a self-contained code cell as required by the subtask.



````python
def modular_exponentiation(base, exp, mod):
    """Calculates (base^exp) % mod using the binary exponentiation algorithm."""
    if mod == 1:
        return 0
    result = 1
    base = base % mod
    while exp > 0:
        if exp % 2 == 1:
            result = (result * base) % mod
        base = (base * base) % mod
        exp //= 2
    return result

# Test the function
b, e, m = 5, 117, 19
print(f"({b}^{e}) % {m} = {modular_exponentiation(b, e, m)}")

# Additional test case
b2, e2, m2 = 2, 10, 1000
print(f"({b2}^{e2}) % {m2} = {modular_exponentiation(b2, e2, m2)}")
````

## Prime and Factorization Algorithms

### Subtask:

**Reasoning**:
Implement a function to find all prime factors of a given integer. This involves dividing the number by 2 until it's odd, then checking odd divisors up to the square root of the remaining number to identify and extract prime factors efficiently.

**Reasoning**:
Implement a function to find all prime factors of an integer using the trial division method, starting with 2 and then checking odd divisors up to the square root of the number, ensuring the code is self-contained and includes test cases.



````python
def get_prime_factors(n):
    """Finds all prime factors of an integer using trial division."""
    n = abs(n)
    factors = []
    if n <= 1:
        return factors

    # Divide by 2 until n is odd
    while n % 2 == 0:
        factors.append(2)
        n //= 2

    # Divide by odd numbers starting from 3
    divisor = 3
    while divisor * divisor <= n:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 2

    # If n is still greater than 2, then n is prime
    if n > 2:
        factors.append(n)

    return factors

# Test the function
test_nums = [56, 100, 315, 97]
for num in test_nums:
    print(f"Prime factors of {num}: {get_prime_factors(num)}")
````

**Reasoning**:
Implement an optimized primality test function that checks for divisors up to the square root of $n$. This approach handles edge cases (like $n \le 1$) and uses $2$ and $3$ as initial checks to speed up the loop by skipping even numbers.

**Reasoning**:
Implement the optimized primality test function as described in the previous reasoning block, checking divisors up to the square root of n and handling small primes as special cases for efficiency.



````python
import math

def is_prime_optimized(n):
    """Checks if a number is prime using an optimized O(sqrt(n)) algorithm."""
    if n <= 1:
        return False
    if n <= 3:
        return True

    # Skip even numbers and multiples of 3
    if n % 2 == 0 or n % 3 == 0:
        return False

    # Check divisors from 5 up to sqrt(n) using the 6k +/- 1 rule
    limit = int(math.sqrt(n))
    for i in range(5, limit + 1, 6):
        if n % i == 0 or n % (i + 2) == 0:
            return False

    return True

# Test the function
test_prime_nums = [2, 3, 17, 19, 25, 29, 97, 100]
for num in test_prime_nums:
    print(f"Is {num} prime? {is_prime_optimized(num)}")
````

**Reasoning**:
Implement the Sieve of Eratosthenes algorithm to efficiently find all prime numbers up to a given limit $n$. This algorithm works by creating a boolean array of size $n+1$ and iteratively marking the multiples of each prime number starting from 2 as composite.

**Reasoning**:
Implement the Sieve of Eratosthenes algorithm in a code cell to find all primes up to a specified limit, providing a list of primes as the result and including a test case for validation.



````python
def sieve_of_eratosthenes(n):
    """Finds all primes up to n using the Sieve of Eratosthenes algorithm."""
    if n < 2:
        return []

    primes = [True] * (n + 1)
    primes[0] = primes[1] = False

    p = 2
    while (p * p <= n):
        if primes[p]:
            # Marking multiples of p as False starting from p*p
            for i in range(p * p, n + 1, p):
                primes[i] = False
        p += 1

    return [i for i, is_prime in enumerate(primes) if is_prime]

# Test the function
limit = 50
print(f"Primes up to {limit}: {sieve_of_eratosthenes(limit)}")
````

Implement a function to find twin primes in a given range. Twin primes are pairs of primes that differ by 2 (e.g., (3, 5)). The function will iterate through the range and verify if both $n$ and $n+2$ are prime, incorporating an optimized primality test to remain self-contained.

**Reasoning**:
Implement a function to find twin primes in a specified range by checking each number and its successor with a difference of two for primality, including an internal primality test to ensure the cell is self-contained.



````python
def find_twin_primes(start, end):
    """Finds all twin prime pairs within a given range [start, end]."""
    def is_prime_internal(n):
        """Internal optimized primality test for self-containment."""
        if n <= 1:
            return False
        if n <= 3:
            return True
        if n % 2 == 0 or n % 3 == 0:
            return False
        i = 5
        while i * i <= n:
            if n % i == 0 or n % (i + 2) == 0:
                return False
            i += 6
        return True

    twin_primes = []
    # We check each number n in the range such that n+2 is also <= end
    for n in range(start, end - 1):
        if is_prime_internal(n) and is_prime_internal(n + 2):
            twin_primes.append((n, n + 2))
    return twin_primes

# Test the function
range_start, range_end = 1, 100
print(f"Twin primes between {range_start} and {range_end}:")
print(find_twin_primes(range_start, range_end))
````

**Reasoning**:
Implement the `modular_pow(base, exponent, modulus)` function using binary exponentiation (also known as the square-and-multiply algorithm). This approach is highly efficient for large exponents, as it reduces the number of multiplications to $O(\log(\text{exponent}))$ and keeps intermediate results small by applying the modulo operation at each step.

**Reasoning**:
Implement the `modular_pow` function using the square-and-multiply algorithm to efficiently compute modular exponentiation, following the specific logic outlined in the subtask instructions.



````python
def modular_pow(base, exponent, modulus):
    """Calculates (base^exponent) % modulus using binary exponentiation."""
    # Handle edge case for modulus 1
    if modulus == 1:
        return 0

    result = 1
    base = base % modulus

    while exponent > 0:
        # If exponent is odd, multiply result with current base
        if exponent % 2 == 1:
            result = (result * base) % modulus

        # Square the base and halve the exponent
        base = (base * base) % modulus
        exponent //= 2

    return result

# Test cases
test_cases = [
    (2, 10, 1000),      # 2^10 % 1000 = 1024 % 1000 = 24
    (5, 117, 19),       # 5^117 % 19 = 1
    (1234567, 890123, 1000000), # Large base and exponent
    (10, 0, 7),         # Power of 0
    (3, 13, 1)          # Modulus 1 case
]

for b, e, m in test_cases:
    print(f"({b}^{e}) % {m} = {modular_pow(b, e, m)}")
````

## Combinatorics: nCr and nPr

### Subtask:
Implement the nCr (Combinations) and nPr (Permutations) algorithms using factorial logic in separate, self-contained code cells.


**Reasoning**:
I will implement the `ncr(n, r)` function in a self-contained code cell, including an internal factorial helper function and validation logic as per the instructions.



````python
def ncr(n, r):
    """Calculates combinations (nCr) using factorial logic, self-contained."""
    def factorial(num):
        if num == 0 or num == 1:
            return 1
        res = 1
        for i in range(2, num + 1):
            res *= i
        return res

    if n < 0 or r < 0 or r > n:
        return "Error: Invalid inputs (ensure n >= r >= 0)"

    # nCr = n! / (r! * (n-r)!)
    return factorial(n) // (factorial(r) * factorial(n - r))

# Test cases
print(f"5C2 = {ncr(5, 2)}")
print(f"10C3 = {ncr(10, 3)}")
print(f"5C5 = {ncr(5, 5)}")
print(f"5C0 = {ncr(5, 0)}")
print(f"3C5 = {ncr(3, 5)}")
````

**Reasoning**:
I will implement the `npr(n, r)` function in a self-contained code cell, including an internal factorial helper function and validation logic as per the instructions.



````python
def npr(n, r):
    """Calculates permutations (nPr) using factorial logic, self-contained."""
    def factorial(num):
        if num == 0 or num == 1:
            return 1
        res = 1
        for i in range(2, num + 1):
            res *= i
        return res

    if n < 0 or r < 0 or r > n:
        return "Error: Invalid inputs (ensure n >= r >= 0)"

    # nPr = n! / (n-r)!
    return factorial(n) // factorial(n - r)

# Test cases
print(f"5P2 = {npr(5, 2)}")
print(f"10P3 = {npr(10, 3)}")
print(f"5P5 = {npr(5, 5)}")
print(f"5P0 = {npr(5, 0)}")
print(f"3P5 = {npr(3, 5)}")
````

**Reasoning**:
Implement the `harmonic_sum(n)` function using a loop and basic arithmetic to calculate the sum of the Harmonic series, including various test cases as specified.



````python
def harmonic_sum(n):
    """Calculates the sum of the first n terms of the Harmonic series."""
    if n <= 0:
        return 0.0

    total_sum = 0.0
    for i in range(1, n + 1):
        total_sum += 1 / i
    return total_sum

# Test cases
test_values = [1, 5, 10, 0, -1]
for val in test_values:
    print(f"Harmonic sum for n={val}: {harmonic_sum(val)}")
````

## Taylor Series: Sin(x)

### Subtask:
Implement Task 24 to calculate $\sin(x)$ using its Taylor series expansion in a self-contained code cell.

**Reasoning**:
Implement the `sin_taylor(x, n)` function to approximate the sine of $x$ (in radians) using the Taylor series expansion: $\sin(x) \approx \sum_{i=0}^{n-1} \frac{(-1)^i x^{2i+1}}{(2i+1)!}$. The implementation will include an internal factorial function and will compare results with Python's `math.sin` to demonstrate accuracy.

**Reasoning**:
Implement the `sin_taylor(x, n)` function to calculate the sine of a given angle in radians using the Taylor series expansion, including an internal factorial function and comparative test cases.



````python
import math

def sin_taylor(x, n_terms=10):
    """Calculates sin(x) using Taylor series expansion: sum of (-1)^i * x^(2i+1) / (2i+1)!"""
    def factorial(n):
        res = 1
        for i in range(2, n + 1):
            res *= i
        return res

    # Normalize x to be within [-pi, pi] for better convergence
    x_normalized = x % (2 * math.pi)
    if x_normalized > math.pi:
        x_normalized -= 2 * math.pi

    sin_x = 0
    for i in range(n_terms):
        # Term: (-1)^i * x^(2*i + 1) / (2*i + 1)!
        power = 2 * i + 1
        term = ((-1)**i * (x_normalized**power)) / factorial(power)
        sin_x += term
    return sin_x

# Test cases
test_angles = [0, math.pi/6, math.pi/4, math.pi/2, math.pi]
print(f"{'Angle (rad)':<15} | {'Taylor Sin':<15} | {'Math Sin':<15} | {'Difference':<15}")
print("-" * 65)
for angle in test_angles:
    taylor_val = sin_taylor(angle)
    math_val = math.sin(angle)
    print(f"{angle:<15.6f} | {taylor_val:<15.6f} | {math_val:<15.6f} | {abs(taylor_val - math_val):<15.6e}")
````

## Taylor Series: e^x

### Subtask:
Implement Task 25 to calculate $e^x$ using its Taylor series expansion in a self-contained code cell.

**Reasoning**:
Implement the `exp_taylor(x, n)` function to approximate the exponential of $x$ using the Taylor series expansion: $e^x \approx \sum_{i=0}^{n-1} \frac{x^i}{i!}$. The function will include an internal factorial helper to maintain self-containment and will be tested against Python's `math.exp` to verify precision.

**Reasoning**:
Implement the `exp_taylor(x, n)` function to calculate the exponential of a number using the Taylor series expansion, including an internal factorial function and comparison with `math.exp` for validation.



````python
import math

def exp_taylor(x, n_terms=20):
    """Calculates e^x using Taylor series expansion: sum of x^i / i!"""
    def factorial(n):
        if n == 0 or n == 1:
            return 1
        res = 1
        for i in range(2, n + 1):
            res *= i
        return res

    exp_x = 0.0
    for i in range(n_terms):
        term = (x**i) / factorial(i)
        exp_x += term
    return exp_x

# Test cases
test_values = [0, 1, 2, -1, 0.5]
print(f"{'x':<10} | {'Taylor e^x':<15} | {'Math exp(x)':<15} | {'Difference':<15}")
print("-" * 60)
for val in test_values:
    taylor_val = exp_taylor(val)
    math_val = math.exp(val)
    print(f"{val:<10.2f} | {taylor_val:<15.6f} | {math_val:<15.6f} | {abs(taylor_val - math_val):<15.6e}")
````
